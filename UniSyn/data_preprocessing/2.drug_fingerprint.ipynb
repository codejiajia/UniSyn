```python
drugmap = np.load('D:\\01Paper\\paper4\\code\\code\\UniSyn\\data_preprocessing\\Drug_map.npy',allow_pickle=True)
```

```python
if isinstance(drugmap, np.ndarray) and drugmap.ndim == 0:
    drugmap = drugmap.item()  
```

```python
selected_drugs = smiles[smiles['Drug'].isin(drugmap.keys())]
```

```python
selected_drugs = selected_drugs.set_index('Drug').loc[list(drugmap.keys())].reset_index()
```

```python
selected_drugs.to_csv('D:\\01Paper\\paper4\\code\\code\\UniSyn\\data_preprocessing\\Drug_smiles.csv')
```

```python
drugs=selected_drugs['SMILES'].tolist()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
```

```python
radius = 3  
nBits = 1024 
fingerprints = []
for smiles in drugs:
    mol = Chem.MolFromSmiles(smiles)
    if mol:
        fp = AllChem.GetMorganFingerprintAsBitVect(mol, radius=radius, nBits=nBits)
        fp_array = np.array(fp)  
        fingerprints.append(fp_array)
    else:
        print(f"Invalid SMILES: {smiles}")
```

```python
type(fingerprints)
```

```python
# **转换为 Pandas DataFrame**
df = pd.DataFrame(fingerprints, columns=[f"{i}" for i in range(nBits)])
```

```python
df 
```

```python
df.insert(0, "id", df.index) 
```
