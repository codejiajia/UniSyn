```python
import anndata as ad
import pandas as pd
import numpy as np
import scanpy as sc
```

```python
exp = pd.read_csv("./OmicsExpressionProteinCodingGenesTPMLogp1.csv")
mu = pd.read_csv("./OmicsSomaticMutationsMatrixDamaging.csv")
nv=pd.read_csv("./OmicsCNGene.csv")
```

```python
exp.columns = exp.columns.str.replace(r"\s*\(.*?\)", "", regex=True)
mu.columns = mu.columns.str.replace(r"\s*\(.*?\)", "", regex=True)
nv.columns = nv.columns.str.replace(r"\s*\(.*?\)", "", regex=True)
```

```python
exp_cell_all=exp['Unnamed: 0'].tolist()
mu_cell_all=mu['Unnamed: 0'].tolist()
nv_cell_all=nv['Unnamed: 0'].tolist()
```

```python
exp_column_names = exp.columns.tolist()
mu_column_names = mu.columns.tolist()
nv_column_names = nv.columns.tolist()
```

```python
cellmap = np.load('./Cell_map.npy',allow_pickle=True)
```

```python
# 如果 drugmap 是一个 0D numpy 数组，先转换为字典
if isinstance(cellmap, np.ndarray) and cellmap.ndim == 0:
    cellmap = cellmap.item()  # 取出内部的字典对象
```

```python
cell_list=list(cellmap.keys())
```

```python
len(cell_list)
```

```python
intersection_cell = list(set(exp_cell_all) & set(cell_list)&set(mu_cell_all)&set(nv_cell_all))
```

```python
len(intersection_cell)
```

```python
col_new=pd.read_csv("./gene_list.txt",header=None)
```

```python
col_new.columns = ["gene_ids"]
```

```python
col_new
```

```python
gene_ids = col_new['gene_ids'].tolist()
```

```python
intersection_exp = list(set(exp_column_names) & set(gene_ids))
```

```python
len(intersection_exp)
```

```python
intersection_mu = list(set(mu_column_names) & set(gene_ids))
```

```python
len(intersection_mu)
```

```python
intersection_nv = list(set(nv_column_names) & set(gene_ids))
```

```python
len(intersection_nv)
```

```python
filtered_cell_exp = exp[exp['Unnamed: 0'].isin(intersection_cell)]
filtered_cell_mu = mu[mu['Unnamed: 0'].isin(intersection_cell)]
filtered_cell_nv = nv[nv['Unnamed: 0'].isin(intersection_cell)]
```

```python
filtered_cell_nv 
```

```python
columns_to_keep = ['Unnamed: 0'] + intersection_exp
columns_to_keep_mu = ['Unnamed: 0'] + intersection_mu
columns_to_keep_nv = ['Unnamed: 0'] + intersection_nv
```

```python
filtered_gene = filtered_cell_exp[columns_to_keep]
filtered_gene_mu = filtered_cell_mu[columns_to_keep_mu]
filtered_gene_nv = filtered_cell_nv[columns_to_keep_nv]
```

```python
filtered_gene_nv
```

```python
filtered_gene.reset_index(drop=True, inplace=True)
filtered_gene_mu.reset_index(drop=True, inplace=True)
filtered_gene_nv.reset_index(drop=True, inplace=True)
```
